```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
## importing all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
## independent features

X=pd.DataFrame(dataset['data'],columns = dataset['feature_names'])
X.head()
```

```python
## dependent features
y=dataset['target']

```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=10)
```

```python
X_train.head()
```

```python
## correlation
X_train.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(X_train.corr(),annot=True)
```

## Hyperparameter Tuning

```python
classifier = DecisionTreeRegressor()
```

```python
params = {
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['sqrt','log2']
}
```

```python
params
```

#### Scikit-learn uses negative values for loss functions in GridSearchCV because higher is better by convention.

```python
import warnings
warnings.filterwarnings('ignore')
grid = GridSearchCV(estimator=classifier,cv=5,n_jobs=-1,param_grid=params,scoring='neg_mean_squared_error')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_  # the more negative the better it is
```

```python
## predictions
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

####  MSE --> where lower is better
#### mean_squared_error(y_test, y_pred)  what you get on actual test data

#### r2_score()  it tells you how much variance your model explains (1 is perfect, 0 is baseline, negative is bad)

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
plt.figure(figsize=(10,15))
tree.plot_tree(grid.best_estimator_,filled=True)
```
