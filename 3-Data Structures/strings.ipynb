```python
# all in build string functions in python"

# 1)len
str = 'hello world'
print(len(str))

# 2)title
str = 'hello world'
print(str.title())

# 3)capitalize
str = 'hello world'
print(str.capitalize())

# 4)upper
str = 'hello world'
print(str.upper())

# 5)lower
str = 'HELLO WORLD'
print(str.lower())

# 6)count  --> it returns the number of occurrences of the substring
str = 'hello world'
print(str.count('o')) 

# 7)find  --> it returns the index of the first occurrence of the substring
str = 'hello world'
print(str.find('o')) 
print(str.find('x')) 

# 8)index  --> it returns the index of the first occurrence of the substring
str = 'hello world'
print(str.index('o'))
# print(str.index('x'))  # it will raise an error if the substring is not found


# 9)endswith
str = 'hello world'
print(str.endswith('world'))  
print(str.endswith('hello'))

# 10)startswith
str = 'hello world'
print(str.startswith('hello'))
print(str.startswith('world'))

# 11)isalnum
str = 'hello123'
print(str.isalnum())  # True
str = 'hello 123'
print(str.isalnum())  # False, because of space


# 12)isspace
str = '   '
print(str.isspace())  # True
str = 'hello'
print(str.isspace())  # False, because it contains characters


# 13)islower
str = 'hello'
print(str.islower())  # True
str = 'Hello'
print(str.islower())  # False, because it contains uppercase character


# 14) isupper
str = 'HELLO'
print(str.isupper())  # True
str = 'Hello'
print(str.isupper())  # False, because it contains lowercase character

# 15)istitle
str = 'Hello World'
print(str.istitle())  # True
str = 'hello world'
print(str.istitle())  # False, because it does not start with uppercase character


# 16)rstrip
str = 'hello world   '
print(str.rstrip())  # it removes trailing spaces AT THE RIGHT END


# 17)lstrip
str = '   hello world'
print(str.lstrip())  # it removes leading spaces AT THE LEFT END

# 18)replace
str = 'hello world'
print(str.replace('world', 'python'))  # it replaces the first occurrence of the substring with the new substring

# 19)join
str = 'hello'
list = ['world', 'python']
print(str.join(list))  # it joins the list elements with the string as separator


# 20)split
str = 'hello world'
print(str.split())  # it splits the string into a list of words
```
