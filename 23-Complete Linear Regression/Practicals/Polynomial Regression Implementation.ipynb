```python
##import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
%matplotlib inline
```

```python
#quadratic equation used- y=0.5x^2+1.5x+2+outliers

x = 6 * np.random.rand(100, 1) - 3
y = 0.5 * x**2 + 1.5*x + 2 + np.random.randn(100, 1)
pt.scatter(x, y,color = 'g')
pt.xlabel('X dataset')
pt.ylabel('y dataset')
pt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)
```

```python
## Lets implement Simple Linear Regression

from sklearn.linear_model import LinearRegression
regression = LinearRegression(n_jobs=-1)
```

```python
regression.fit(x_train,y_train)
```

```python

from sklearn.metrics import r2_score
score = r2_score(y_test,regression.predict(x_test))
print(score)
```

```python
## Lets visualize this model

pt.plot(x_train,regression.predict(x_train), color= 'r')
pt.scatter(x_train,y_train)
pt.xlabel('x dataset')
pt.ylabel('y dataset')
pt.show()
```

#### Hence from the above model training which is done using simple linear regression model it is observd that we are not getting a good ml model on the basis of r2 score.

#### hence for this polynomial regression we will use the concept of degree and include bias, this include bias will determine the slope value where we will continously change it manually at last we will use the concept of pipeline to make a generalize formula to predict degree and create plot for non linear dataset

#### just like in linear regression we are doing standardization in on independent/ input features i.e is x train and x test values. similarly we will do that step in polynomial regression where we will will polynomailly transform it using the independent/ input features i.e x train and x test vales 

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=True)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

#### in polynomial regression we simply use the steps just like we are solving the linear regression model. As with degree 0 we will get a constant and with degree 1 we will get the simple linear regression moidel

```python
# traing the model on the basis of polynomial datset
from sklearn.linear_model import LinearRegression 
regression = LinearRegression(n_jobs=-1)
regression.fit(x_train_poly,y_train)

#testing the model 
y_pred = regression.predict(x_test_poly)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score)
```

## Pipeline Concepts 

```python
from sklearn.pipeline import Pipeline
def poly_regression(degree):
    x_new = np.linspace(-3,3,200).reshape(200,1)

    poly_features = PolynomialFeatures(degree = degree, include_bias=True)
    lin_reg = LinearRegression()
    poly_regression = Pipeline([
        ('poly_features',poly_features),
        ('lin_reg',lin_reg)
    ])
    poly_regression.fit(x_train,y_train)
    y_pred_new = poly_regression.predict(x_new)

    pt.plot(x_new,y_pred_new,'r',label='degree'+str(degree),linewidth = 2)
    pt.plot(x_train,y_train,'b.',linewidth = 3)
    pt.plot(x_test,y_test,'g.',linewidth = 3)
    pt.legend(loc='upper left')
    pt.xlabel('x')
    pt.ylabel('y')
    pt.show()
```
