```python
1) probability distribution gives us the probability distribution of values in the random variable
2) the random value can have two types of distribution: discrete or continous
3) to measure tthe the probabiility distriubution of discrete values we use the probability mass function (pmf)
4) to measure the probability distribution of continuous values we use the probability density function (pdf)
5) the cumulative distribution function (cdf) gives us the probability that a random variable is less than or equal to a certain value

# PMF important  points
1)PMF â€“ Probability Mass Function
    In-depth Meaning:
    PMF describes the probability distribution of a discrete random variable.
    For a random variable X, PMF gives P(X=x), i.e., the probability that X takes a specific value x. It generally means we will get direct probabilities for specific outcomes.
    It works only for discrete values (you can count them, like 0, 1, 2...).

ðŸ“ˆ Visual Representation:
Imagine a bar graph:
Each bar height = probability of that discrete value.

ðŸ§ª Example:
Toss a fair coin twice:
Possible outcomes (X = number of heads): 0, 1, 2

P(X=1) = 1/6

P(X=2) = 1/6 and so on...

âœ… Properties:
ð‘ƒ(X=x) >= 0   #--> this property says that we will get the proibability of a discrete point which will never be 0 

âˆ‘ P(X=x) = 1   #--> this property says that the sum of all probabilities will be equal to 1

while constructing a PMF graph, we can use the following steps:
1. **Identify the Random Variable**: Determine the discrete random variable you are interested in.
2. **List Possible Outcomes**: Enumerate all possible values the random variable can take.
3. here we have parameters like p,q,k. k is the outcome if k=1 then its success or its 0 then its failure, p is the probability of success, and q is the probability of failure.
4. **Calculate Probabilities**: Use the PMF formula to calculate the probability for each outcome.
    the PMF formula is given by:
    pmf = p^k * (1-p)^(1-k)   [this is for bernouli distribution]
5. **Plot the Graph**: Use a plotting library (like Matplotlib in Python) to visualize the PMF by plotting the outcomes against their probabilities.




# PDF important points
1) PDF â€“ Probability Density Function
    In-depth Meaning:
    PDF is used for continuous variables (uncountably infinite values like time, height, weight).
    The PDF does not give a probability directly. Instead, it gives a density, and you get probabilities by finding the area under the curve between two points.

    key formula - p(a < X < b) = âˆ«[a to b] f(x) dx
    It works for continuous values (you can't count them, like 0.1, 0.2, 0.3...).

    ðŸ§ª Example:
    Suppose human heights follow a normal distribution with mean 170 cm and standard deviation 10 cm:

    The PDF tells you how â€œlikelyâ€ any height is, but the actual probability of being exactly 170 cm = 0.
    Instead, you'd ask: What's the probability of being between 160 and 180 cm? â†’ That's an area under the curve.

    ðŸ“ˆ Visual:
    A smooth curve (bell curve for normal distributions).

    The total area under the PDF = 1.

âœ… Properties:
    ð‘“(ð‘¥) â‰¥ 0   #--> this property says that the probability density function will never be negative

    âˆ«[âˆ’âˆž to âˆž] f(x) dx = 1   #--> this property says that the total area under the curve will always be equal to 1

    P(X=a) = 0 for any single point   #--> this property says that the probability of a continuous random variable taking on any single value is always zero

while constructing a PDF graph, we can use the following steps:
1. **Choose a Distribution**: Decide on the type of distribution (e.g., normal, uniform, exponential).
2. **Define Parameters**: Set the parameters for the distribution (e.g., mean and standard deviation for a normal distribution).
3. **Generate Data Points**: Create a range of x-values (e.g., from -3 to 3 for a standard normal distribution).
4. **Calculate PDF Values**: Use the PDF formula to calculate the corresponding y-values for each x-value. 
5. **Plot the Graph**: Use a plotting library (like Matplotlib in Python) to visualize the PDF by plotting x-values against y-values.



# CDF important points
1)  Used for: Both Discrete and Continuous Random Variables
2) CDF â€“ Cumulative Distribution Function
    In-depth Meaning:
    CDF gives the probability that a random variable X is less than or equal to a certain value x.
    For a random variable X, CDF is defined as F(x) = P(X â‰¤ x).
    It works for both discrete and continuous variables.

    ðŸ§ª Example:
    For a discrete random variable (like rolling a die):
    - P(X â‰¤ 3) = P(X=1) + P(X=2) + P(X=3)
    
    For a continuous random variable (like height):
    - F(170 cm) = P(X â‰¤ 170 cm)

    ðŸ“ˆ Visual:
    A smooth, increasing curve that starts at 0 and ends at 1.

âœ… Properties:
    F(x) is non-decreasing   #--> this property says that the CDF will never decrease as x increases

    lim (x â†’ -âˆž) F(x) = 0   #--> this property says that as x approaches negative infinity, the CDF approaches 0

    lim (x â†’ âˆž) F(x) = 1   #--> this property says that as x approaches positive infinity, the CDF approaches 1

    P(a < X â‰¤ b) = F(b) - F(a)   #--> this property says that the probability of X being between a and b can be calculated using the CDF


âš”ï¸ PMF vs PDF vs CDF â€“ Side-by-Side

| Feature       | PMF (Probability Mass Function)  | PDF (Probability Density Function)   | CDF (Cumulative Distribution Function)   |
|---------------|----------------------------------|------------------------------------- |------------------------------------------|
| Type          | Discrete                         | Continuous                           | Both                                     |
| Definition    | P(X=x) for discrete X            | f(x) for continuous X                | F(x) = P(X â‰¤ x)                          |
| Probability   | Direct for specific values       | Density, area gives probability      | Cumulative probability                   |
| Graph         | Bar graph                        | Smooth curve                         | Non-decreasing curve                     |
| Total Area    | Sum = 1                          | Area under curve = 1                 | Not applicable                           |
|integration    | Not applicable                   | âˆ«[âˆ’âˆž to âˆž] f(x) dx = 1               | Not applicable                           |
|value at point | P(X=a) > 0 for discrete X        | P(X=a) = 0 for continuous X          | F(x) gives cumulative probability        |


```
