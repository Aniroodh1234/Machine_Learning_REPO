## Naive Baye's Machine Learning Algorithms Implementation

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y=load_iris(return_X_y=True)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred=gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
df.drop('size',axis=1,inplace=True)
```

```python
X = df[['total_bill', 'tip', 'sex','day','time']]
y = df['smoker']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.10,random_state=10)
```

```python
## performing label encoding 
from sklearn.preprocessing import LabelEncoder
l1= LabelEncoder()
l2= LabelEncoder()

X_train['sex'] = l1.fit_transform(X_train['sex'])
X_train['time'] = l2.fit_transform(X_train['time'])

X_train.head()
```

```python
X_test['sex'] = l1.transform(X_test['sex'])
X_test['time'] = l2.transform(X_test['time'])
X_test.head()
```

#### whenever we are performing column transformations on multiple column and using different techniques we use column transformation

in this case we are using column transformation with single transformation technique and that is one hot encoding. Further in random forest we have used multiple transformation techinque using column transformation 

```python
## performing one hot encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(transformers = [('onehot',OneHotEncoder(drop='first'),[3])],remainder='passthrough')
```

```python
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)

X_train = ct.fit_transform(X_train)
```

```python
X_train
```

```python
# Recreate X_test from the original DataFrame to ensure it has the correct columns and types
X_test = df[['total_bill', 'tip', 'sex', 'day', 'time']].iloc[y_test.index]

# Apply label encoding to 'sex' and 'time' columns
X_test['sex'] = l1.transform(X_test['sex'])
X_test['time'] = l2.transform(X_test['time'])

# Now apply the ColumnTransformer
X_test = ct.transform(X_test)
X_test
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(X_train, y_train)
```

```python
y_pred =gnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
