## Support Vector Classifier Implementation

in support vector machines whether its classification or regression our main aim is to draw a best fit line along with the marginal line and the distance between the two magrinal line must be maximum. In SVM the best fit line would be constructed where there is best possible separation in the datapoints. 

The best fit lines are drawn on the basis of svm kernel if we take kernel as linear. It means datpoints are are linear and we would draw a straight best fit line. If we set it as rbg, sygmoid, polyomial it means datapoints are non linear and. Hence, it will convert them to a higher state by applying some maths transformation and make it separable datpoints so that we can make best fit line.

Among all this we can also the apply the concept of hyper parameter tunning to increase the model accuracy

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Lets create synthetic data points
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

The coefficients are only be fetched in linear svm kernles 

```python
svc.coef_
```

```python
## Prediction
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

## Hyperparametere Tuning With SVC

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
model = SVC()
param = {
    'C':[0.1,1,10,100,1000],
    'gamma':[1, 0.1, 0.01, 0.001, 0.0001],
    'kernel':['rbf']
}
```

```python
grid = GridSearchCV(estimator=model,param_grid=param,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
